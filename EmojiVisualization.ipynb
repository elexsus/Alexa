```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
path='/content/emoji.csv'
emoji_data=pd.read_csv(path)
emoji_data.head()
```

```
emoji_data.tail()
```

```
emoji_data['group'].value_counts().plot.pie(figsize=(10,8))  
# show plot 
plt.show()
```

```
emoji_objects=emoji_data[emoji_data.group=='Objects']
emoji_objects['sub_group'].value_counts().plot.pie(figsize=(10,8))  
# show plot 
plt.show()
```

```
plt.figure(figsize=(16,10))
sns.countplot('group',data=emoji_data)
plt.show()
```

```
plt.figure(figsize=(16,10))
emoji_animal=emoji_data[emoji_data.group=='Animals & Nature']
sns.countplot('sub_group',data=emoji_animal)
plt.show()
```
